```python
import os
import shutil

source = "transformed-images"
cat_dest = "dataset/cat"
dog_dest = "dataset/dog"

for file in os.listdir(source):
    if file.lower().startswith("cat"):
        shutil.copy(os.path.join(source, file), cat_dest)
        
    elif file.lower().startswith("dog"):
        shutil.copy(os.path.join(source, file), dog_dest)

print("✅ Images copied successfully!")
```

```python
import os

print("Cats:", len(os.listdir("dataset/cat")))
print("Dogs:", len(os.listdir("dataset/dog")))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

img_size = 150

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_data = datagen.flow_from_directory(
    "dataset",
    target_size=(img_size, img_size),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_data = datagen.flow_from_directory(
    "dataset",
    target_size=(img_size, img_size),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),

    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=5
)
```

```python
model.save("cnn_model.h5")
```
